```python
high_rating = df[df["rating"] > 1]
high_rating.head()
```

# Quantity column was not available in dataset,
# so a default quantity value of 1 was used.

```python
df["quantity"] = 1
df["total_amount"] = df["final_price"] * df["quantity"]
```

```python
df.groupby("category")["rating"].mean()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
