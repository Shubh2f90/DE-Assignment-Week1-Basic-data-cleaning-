```python
# checking which columns have null values
df.isnull().sum()
```

```python
df['discount'] = df['discount'].fillna(df['discount'].median())

df['what_customers_said'] = df['what_customers_said'].fillna('No feedback')

df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

df['variations'] = df['variations'].fillna('No Variations')
print(df.isnull().sum())
```

```python
df['final_price'] = df['final_price'].astype(str)

df['final_price'] = df['final_price'].str.replace('₹', '', regex=False)

df['final_price'] = df['final_price'].str.replace(',', '', regex=False)

df['final_price'] = df['final_price'].str.replace('"', '', regex=False)

# convert to numeric

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Verify datatype
df['final_price'].dtype
```

```python
print("Before:", len(df))

df = df.drop_duplicates()

print("After:", len(df))

print("Duplicate rows:", df.duplicated().sum())
```

```python
high_rating = df[df["rating"] > 4]
high_rating.head()
```

```python
# Finding Total Amount
df["quantity"] = 1
df["total_amount"] = df["final_price"] * df["quantity"]

# Discount
df['discount_amount'] = df['initial_price'] - df['final_price']

#Show
df[['initial_price',
    'final_price',
    'quantity',
    'discount_amount',
    'total_amount']].head(10)
```

```python
seller_products = df.groupby('seller_name')['product_id'].count().sort_values(ascending=False)

seller_products.head(10)
```

```python
category_discount = df.groupby('category')['discount'].mean().sort_values(ascending=False)

category_discount.head(10)
```

```python
# clean text columns

text_columns = ['seller_name', 'category']

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.encode('ascii', 'ignore')
        .str.decode('ascii')
        .str.strip()
    )

print("Text columns cleaned successfully")
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
print("Cleaned dataset saved successfully")
```
